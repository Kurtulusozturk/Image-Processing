```python
#Library 
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
```

```python
img = load_img("Dataset/a.jpg")
"""
İlk olarak veri tipini ayarlamamız lazım 
1.(100,100,3)
2.PIL --> array
3.(1,100,100,3)1 görüntü sayısı

"""
img_array = img_to_array(img)
print("img_array",img_array.shape)

reshaped_img_array = img_array.reshape((1,) + img_array.shape)
print("reshaped_img_array",reshaped_img_array.shape)
```

```python
data_generator = ImageDataGenerator(rotation_range=65,
                                    width_shift_range=0.2,
                                    height_shift_range=0.2,
                                    zoom_range=0.2,
                                    horizontal_flip=True,
                                    fill_mode='nearest')
```

```python
counter = 0
for i in data_generator.flow(reshaped_img_array,
                             batch_size=1,
                             save_to_dir="Dataset/augmented_yoda",
                             save_prefix="jedi",
                             save_format="jpeg"):
    counter = counter+1
    if counter == 25:
        break
```

```python
data_generator_directory = ImageDataGenerator(rotation_range=65,
                                                width_shift_range=0.2,
                                                height_shift_range=0.2,
                                                zoom_range=0.2,
                                                horizontal_flip=True,
                                                vertical_flip=True,
                                                fill_mode='nearest')
```

```python
counterx = 0
for i in data_generator_directory.flow_from_directory("Dataset/yodaclone",
                                                         batch_size=1,
                                                         save_to_dir="Dataset/augmented_yodaclone",
                                                         save_format="jpeg"):
    counterx = counterx+1
    print("Processing... (",counterx,"/200)")
    if counterx == 200:
        print("Done!")
        break
```
