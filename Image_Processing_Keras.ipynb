```python
#Library

from tensorflow.keras.preprocessing.image import load_img
```

```python
#reading/loading image
img = load_img("Dataset/a.jpg")#PIL ÅŸeklinde resim okur

print("Type:",type(img))
print("Format:",img.format)
print("Size:",img.size)
print("Mode:",img.mode,end = "\n\n")

print("All properties:",img)
```

```python
#displaying image
import matplotlib.pyplot as plt

img = load_img("Dataset/a.jpg")
img1 = load_img("Dataset/b.jpg")

plt.imshow(img)
plt.show()
plt.imshow(img1)
plt.show()
```

```python
#PIL --> numpy array
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img

img = load_img("Dataset/a.jpg")
print("Type first:",type(img))

img = img_to_array(img)
print("Type second:",type(img))

img = array_to_img(img)
print("Type last:",type(img))
```
